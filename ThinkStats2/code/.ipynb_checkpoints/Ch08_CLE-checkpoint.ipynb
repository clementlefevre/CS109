```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
import math as math
import nsfg
import random
```

```python
mean =0
std = 1
n = 7
m =1000
x_mean =[]
median =[]

for _ in range(m):
    xs = np.random.normal(mean,std,n)
    x_mean.append(np.mean(xs))
    median.append(np.median(xs))
    
def MSE(estimates,actual):
    return math.sqrt(np.mean([(estimate-actual)**2 for estimate in estimates]))

print MSE(x_mean,mean)
print MSE(median,mean)
```

```python
# Chimpanzee population weight

mean = 90
sigma = 7.5

m = 1000
n = 7 

x_mean = []

for _ in range(m):
    
    xs = np.random.normal(mean,sigma,7)
    x_mean.append(xs.mean())

x_mean_array  = np.asarray(x_mean)
x_mean_array.sort()

cdf = np.cumsum(x_mean_array)/x_mean_array.sum()


# plot the results
#plt.plot( x_mean_array, kde(x_mean_array) )

#plt.plot(x_mean_array,pdf)
# plt.plot(x_mean_array,estimate)
plt.plot(x_mean_array,cdf)
plt.axvline(np.percentile(x_mean_array,5))
plt.axvline(np.percentile(x_mean_array,95))

#compute Root Mean Square Error :

RMSE= math.sqrt(np.sum((x_mean_array-mean)**2)/m)

print RMSE
```

```python
mean = 0
sigma = 1
m = 1000
n =7000
x_mean=[]
x_median = []

#compute mean error on mean and median
for _ in range(m):
    xs = np.random.normal(mean,sigma,n)
    x_mean.append(xs.mean())
    x_median.append(np.median(xs))

    
def MeanError(x_array):
    return np.mean(np.asarray(x_array)-mean)

plt.hexbin(x_mean,x_median)


print MeanError(x_mean)
print MeanError(x_median)
```

```python
def lambada(n):
    lambdo = 2
    
    m = 1000
    x_mean=[]

    for _ in range(m):
        xs = np.random.exponential(1.0/lambdo,n)
        x_mean.append(1.0/xs.mean())
    x_mean.sort()  

    plt.plot(x_mean,stats.expon.cdf(x_mean))
    RMSE = math.sqrt(np.sum((np.asarray(x_mean)-lambdo)**2)/m)
    print ("RMSE" , RMSE)
    plt.axvline( np.percentile(x_mean,5))
    plt.axvline(np.percentile(x_mean,95))
    return RMSE

```

```python
lambada(10)
```

```python
lambada(10000)
```
