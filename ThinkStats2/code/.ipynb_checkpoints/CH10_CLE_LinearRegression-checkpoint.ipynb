```python
from __future__ import division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
import math as math
import nsfg
import random
import first
import thinkstats2
import statsmodels.formula.api as smf
```

## Compute manually the OLS (Ordinary Least Squares) parameters

```python
def compute_beta(x,y):
    return np.sum((x-x.mean())*(y-y.mean()))/np.sum((x-x.mean())**2)

def compute_alpha(x,y,beta):
    return y.mean()-beta*x.mean()

live,firsts,others = first.MakeFrames()
x, y = live.agepreg, live.totalwgt_lb
beta = compute_beta(x,y)
alpha = y.mean()-beta*x.mean()
plt.hexbin(x,y)
plt.plot(x,alpha + beta*x, color ='r')
plt.ylabel('totalwgt_lb')
plt.xlabel('agepreg')
print alpha,beta


def ols(x):
    return alpha + beta *x
```

## Group by quantiles and evaluate the residuals

```python
bins = [0,15,20,25,30,35,40,45,60]
labels = [15,20,25,30,35,40,45,60]
live['categories_age'] = pd.cut(live.agepreg,bins,labels = labels)
live = live[['agepreg','totalwgt_lb','categories_age']]
live['residual'] = ols(live.agepreg)-live.totalwgt_lb
live.head()
groupy = live.groupby('categories_age')

# compute the residual for each age category for each quantiles

quantiles = [25,50,75]

# live = live.sort_values(by='totalwgt_lb')
live = live.dropna()
live = live.reset_index(drop=True)
q25 = groupy.quantile(.25)
q50 = groupy.quantile(.5)
q75 = groupy.quantile(.75)

quantiles = [q25,q50,q75]

mean_age = groupy.agepreg.mean()

for q in quantiles:
    
    plt.plot(mean_age,-q.residual)

plt.xlabel("age category")
plt.ylabel('residual')
    
```

## Estimation

```python
#sample live rows and compute the corresponding intercept and slope

def get_sample(df):
    indices = np.random.choice(live.index, len(live), replace=True)
    sample = live.loc[indices]
    return sample

test =[]
m = 101


def get_data():
    sample = get_sample(live)
    
    beta = compute_beta(sample.agepreg,sample.totalwgt_lb)
    alpha = compute_alpha(sample.agepreg,sample.totalwgt_lb,beta)
    return alpha,beta
    
test = np.asarray([get_data() for _ in range(m)])

print ("intercept : ", test[:,0].mean(),(test[:,0]-beta).std(), "\n", "CI : [",np.percentile(test[:,0],5),":",np.percentile(test[:,0],95),"]" )
print ("slope : ",test[:,1].mean(),(test[:,1]-alpha).std(), "CI : [",np.percentile(test[:,1],5),":",np.percentile(test[:,1],95),"]" )


```

## Plot confidence interval of resampling results

```python
interval = [5,25,75,95]
param_estimation = []

for i in interval:
    param_estimation.append((np.percentile(test[:,0],i),np.percentile(test[:,1],i)))
  
for t in test:
    alpha_interval = t[0]
    beta_interval = t[1]
    #plt.plot(live.agepreg,alpha_interval + beta_interval*live.agepreg,alpha =.3)

#compute the 90% CI lines for each agepreg value:
line_95,line_5,line_25,line_75 =[],[],[],[]

live = live.sort_values(by='agepreg')

for age in live.agepreg:
    ys= test[:,0] +  age* test[:,1]
    line_95.append(np.percentile(ys,95))
    line_5.append(np.percentile(ys,5))
    line_25.append(np.percentile(ys,25))
    line_75.append(np.percentile(ys,75))
    
       
    
plt.plot(live.agepreg,np.asarray(line_95),color='c')
plt.plot(live.agepreg,np.asarray(line_5),color='c')
plt.fill_between(live.agepreg,np.asarray(line_95),np.asarray(line_5),alpha=.3)

plt.plot(live.agepreg,np.asarray(line_75),color='c')
plt.plot(live.agepreg,np.asarray(line_25),color='c')
plt.fill_between(live.agepreg,np.asarray(line_75),np.asarray(line_25),alpha=.6,color='y')

plt.ylabel('totalwgt_lb')
plt.xlabel('age')
plt.title("With 101 Sampled distributions : Linear Parameter variability according to Confidence interval 90% and 50%")
```

##  Goodness to Fit

```python
Residuals = live.totalwgt_lb -ols(live.agepreg)
Residuals  = np.sort(Residuals)
plt.plot(live.agepreg,Residuals)
plt.xlabel("age")
plt.ylabel("Residual")
RMSE = np.std(Residuals)
print "RMSE :",RMSE

R_2 = 1-np.sum(Residuals**2)/np.sum((live.totalwgt_lb-live.totalwgt_lb.mean())**2)
R_2D2 = 1- (np.std(Residuals)/np.std(live.totalwgt_lb))**2
print "R_2",R_2
print "R_2D",R_2D2
```

Using the data from the BRFSS, compute the linear least squares fit for log(weight)

```python
import brfss

df = brfss.ReadBrfss(nrows=None)
df = df.dropna(subset=['htm3', 'wtkg2'])
heights, weights = df.htm3, df.wtkg2
weight_absolute =weights
weights = np.log10(weights)
```

```python
beta = compute_beta(df.htm3,np.log10(df.wtkg2))
alpha =  compute_alpha(df.htm3,np.log10(df.wtkg2),beta)

bins = range(130,280,5)
labels = ["<"+ str(_) for _ in bins[1:]]
df['height_categories'] = pd.cut(df.htm3,bins, labels = labels)
df['residuals']= np.log10(df.wtkg2)-(df.htm3*beta+alpha)


groupy = df.groupby('height_categories')
mean_height = groupy.htm3.mean()

quantiles = [25,50,75]

for q in quantiles:
    plt.plot(mean_height, groupy.residuals.quantile(q/100))
plt.ylabel('weight - residuals')
plt.xlabel('height')
df.boxplot("residuals",by="height_categories",rot=90)

df.head()
```
