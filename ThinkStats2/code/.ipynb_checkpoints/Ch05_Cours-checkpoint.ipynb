```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import nsfg
import numpy as np

%matplotlib inline
```

```python
preg = nsfg.ReadFemPreg()
```

```python
weight = preg.birthwgt_lb.dropna().to_frame()
```

```python
weight = weight.sort_values(by='birthwgt_lb', ascending=True)
size = len(weight)
mean =weight.birthwgt_lb.mean()
std = weight.birthwgt_lb.std()
```

```python
random_weight = np.random.normal(0,1,len(weight))
random_weight_adjusted = np.random.normal(mean,std,size)
random_weight.sort()
random_weight_adjusted.sort()
weight['random_weight']= random_weight
weight['random_weight_adjusted']= random_weight_adjusted
```

```python
weight.tail()
plt.plot(weight.random_weight,weight.birthwgt_lb)
plt.plot(weight.random_weight_adjusted,weight.birthwgt_lb)
```

```python
plt.plot(weight.birthwgt_lb,weight.random_weight)
```

```python
sp.stats.probplot(weight.birthwgt_lb[30:6000], dist="norm", plot=plt)
plt.title("Normal Q-Q plot")
plt.show()
```

```python
df = pd.read_csv('PEP_2012_PEPANNRES_with_ann.csv')

df = df[df['GEO.id']!="Id"]

df.respop72012 = pd.to_numeric(df.respop72012)
# df.info()
pop= df.respop72012.to_frame()
pop = pop.sort_values(by='respop72012')
pop['pop_log10'] = np.log10(pop.respop72012)

pop = pop.reset_index(drop=True)
pop.tail()

pop.pop_log10.plot()
```

```python
def compute_Pareto_CDF(x):
    alpha = 1
    xm = 0.5
    return (1-(x/xm)**(-alpha))

pop['CDF_Pareto'] = pop.pop_log10.apply(lambda x : (1-compute_Pareto_CDF((x))))



pop = pop[pop.respop72012>0]
plt.scatter(pop.pop_log10,pop.CDF_Pareto)
pop.tail()
```
